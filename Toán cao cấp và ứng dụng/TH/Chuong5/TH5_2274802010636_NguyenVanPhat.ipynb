```python
import sympy
from sympy import Derivative, Symbol, solve, pprint, sin, Integral, sqrt, Limit, S
from sympy.plotting import plot
```

```python
x = Symbol("x")
c = Symbol("c")
delta = Symbol("delta")
f = x * sin(1/x)
c = 0
delta = 1/4
print(Limit(f, x, 0, dir = '+'))
print(Limit(f, x, 0).doit())
sympy.plot(f, (x, c - delta, c + delta))
```

```python
x = Symbol("x")
f = -x**2 +4*x - 3
d1 = Derivative(f, x).doit()
cuc_tri = solve(d1)
print(cuc_tri)
# print(d1)

A = cuc_tri[0]
d2 = Derivative(d1, x).doit()
d2.subs({x:A}).evalf()
x_min = 0
x_max = 4
print(" Gia tri ham tai cuc tri: ",f.subs({x:A}).evalf())
print("Gia tri ham tai x_min:",f.subs({x:x_min}).evalf())
print("Gia tri ham tai x_max:",f.subs({x:x_max}).evalf())
print(d2)
plot(f, (x, 0, 4))
```

```python
x = Symbol("x")
a = Symbol("a")
f = -2*x**3 + 2*a*x
d1 = Derivative(f, x).doit()
d2 = solve(d1)
print(f"Cuc tri cua {d1}: {d2}")
x_1 = f.subs({x:0}).evalf()
x_2 = f.subs({x:sqrt(a)}).evalf()
x_3 = f.subs({x:sqrt(a/3)}).evalf()
print(f"x1 = {x_1}")
print(f"x2 = {x_2}")
print(f"x3 = {x_3}")
```

```python
x = Symbol("x")
f = -10*x**2 + 25000*x - 120000000
d = Derivative(f,x).doit()
cuc_tri = solve(d)
print(cuc_tri)

A = cuc_tri[0]
Max = f.subs({x:A}).evalf()
print(f"Loi nhuan toi da:", Max)
```

```python
q = Symbol("q")
f = -0.8*q + 150
print("Dap an:", Integral(f, (q,0,25)).doit().evalf())
#cau 1
d = solve(-0.8*q + 150 - 5.2*q)
print(d)
print("Gia tri la:",f.subs({q:d[0]}).evalf())

#cau 2
tong_tien = Integral(f, (q,0,25)).doit().evalf() - f.subs({q:d[0]}).evalf()
print(f"Tong tien: {tong_tien}")

#cau 3
f1 = 5.2*q
a = 130*25 - Integral(f1, (q,0,25)).doit().evalf()
print("Kq = ",a)
```
